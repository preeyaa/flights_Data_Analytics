# Trying Seaborn

```python
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
#sns.get_dataset_names()
```

## loading flight dataset

```python
flights= sns.load_dataset('flights')
flights.shape # To see total rows and columns
```

```python
flights.head() # To see first 5 rows
```

```python
flights.all #To see complete data
```

## Using pivot table function for viewing differently

```python
flights.pivot_table(index='month',
                    columns='year',
                    values='passengers')
```

## Plotting the data

```python
sns.lmplot(x='year',y='passengers', data=flights)
```

```python
sns.lmplot(x='year', y='passengers', data=flights,
           fit_reg=False, # No regression line
           hue='month')   # Color by month
```

```python
##Use a new color palette
colors = ['#78C850',#Jan
                    '#6890F0',  # Feb
                    '#A8B820',  # March
                    '#A8A878',  # April
                    '#A040A0',  # May
                    '#F8D030',  # June
                    '#808030',  # July
                    '#EE99AC',  # August
                    '#C03028',  # September
                    '#F85888',  # October
                    '#B8A038',  # Nov
                    '#705898',  # Dec
                   ]
sns.lmplot(x='year', y='passengers', data=flights,
           fit_reg=False, # No regression line
           hue='month',   # Color by month
           palette=colors)
```

```python

# Enlarge the Figure
plt.figure(figsize=(10,6))

sns.lmplot(x='year', y='passengers', data=flights,
           fit_reg=False, # No regression line
           hue='month',   # Color by month
           palette=colors)

plt.title('Passengers in Flights from 1949 till 1960')
```

### Conclusion:
Passengers kept increasing as years passed by and the maximum passengers are in the month July and August.
